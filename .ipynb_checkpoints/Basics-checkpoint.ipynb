I'm going to combine both datasets, raising two main questions for this analysis:
 - Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

I'm going to start with TAFE, as it is the easier dataset to work with in this instance.

I'm going to create a new column, that will have either True or False values. If employee filled in either of the contributing factors I'll mark the dissatisfaction as True.

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts(dropna=False)
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(dropna=False)
```

We can see that both columns contain, either "-", NaN, or some other value, this data needs to be converted into True or False values. I think NaN can be left as is, but "-" will be converted into False, and any other value other than NaN or "-" into True.

```python
def convert_tafe_values(val_string):
    if val_string == '-':
        return False
    elif pd.isnull(val_string):
        return np.nan
    else:
        return True
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'] = tafe_resignations['Contributing Factors. Job Dissatisfaction'].apply(convert_tafe_values)
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts(dropna=False)
```

```python
tafe_resignations['Contributing Factors. Dissatisfaction'] = tafe_resignations['Contributing Factors. Dissatisfaction'].apply(convert_tafe_values)
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts(dropna=False)
```

```python
tafe_resignations.insert(4, 'dissatisfied', tafe_resignations[['Contributing Factors. Dissatisfaction','Contributing Factors. Job Dissatisfaction']].any(axis='columns',skipna=False))
tafe_resignations_up = tafe_resignations.copy()
```

 #### Categorizing employees by years of service
 
 After combining the dataset the *'institute_service'* column is a mess. It contains, float, and string values. To continue with analysis I'll have to clean this column up.
 
According to [this article](https://www.businesswire.com/news/home/20171108006002/en/Age-Number-Engage-Employees-Career-Stage) there are four distinct categories, that I could fit all of these employee types:
 - **Newbie**: Employees with 3 years or less of experience at the company.
 - **Sophomore**: Employees with 4-7 years of experience at the company.
 - **Tenured**: Employees with 8-10 years of experience at the company.
 - **Sage**: Employees with 11 years or more of experience at the company.

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)').astype('float')
combined_updated['institute_service'].value_counts()
```

```python
def employee_yos_mapper(years):
    if pd.isnull(years):
        return np.nan
    elif years <= 3:
        return 'Newbie'
    elif (years > 3) & (years < 8):
        return 'Sophomore'
    elif (years > 7) & (years < 11):
        return 'Tenured'
    elif years > 10:
        return 'Sage'
```

```python
combined_updated.insert(3, 'service_cat', combined_updated['institute_service'].apply(employee_yos_mapper))
combined_updated.head(10)
```

#### Performing initial analysis

```python
table = combined_updated.pivot_table(index='service_cat', values='dissatisfied')
table.plot(kind='bar')
plt.show()
```

Can see here, that people who worked the longest 11+ years are exiting the institute due to some kind of dissatisfaction. 
What is notable, that employees who haven't worked there long (less than 3 years) are resigning less due to some dissatisfaction, and more so due to other factors.
